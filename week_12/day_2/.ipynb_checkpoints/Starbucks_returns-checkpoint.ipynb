```python
drinks = (
drinks_content
.loc[:, ["Beverage_category", "Beverage", "Beverage_prep", "Calories"]]
.copy())

```

```python
drinks
```

```python
drinks.isna().sum()
```

```python
drinks["calorie_diff"] = drinks.Calories - 135
```

```python
drinks
```

```python
(drinks
.groupby("Beverage_category")
.Calories.mean()
.reset_index())
```

```python


mean_calories = 193.87
```

```python
(drinks
 .loc[(drinks.Calories > mean_calories)])
```

```python
beverage_prep_cals = (
    drinks
    .groupby('Beverage_prep')
    .Calories.mean()
    .reset_index()
)
```

```python
overall_mean_cals = drinks.Calories.mean()

(
    beverage_prep_cals
    .loc[(beverage_prep_cals.Calories > overall_mean_cals)]
)
```

```python
(drinks
.groupby("Beverage")
.agg({"Calories":"min"})
.reset_index()
.sort_values("Calories", ascending = True)
.loc[drinks.Beverage != "Tazo® Tea"])
```

```python
espresso_drinks =(
drinks
    .loc[drinks.Beverage_category == "Classic Espresso Drinks"])
```

```python
espresso_drinks
```

```python
(espresso_drinks
.groupby("Beverage")
.Calories.mean()
.reset_index())
```
